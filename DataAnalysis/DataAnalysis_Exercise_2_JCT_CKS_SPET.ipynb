**d.** We can also look at this history of $CO_2$ emissions by each country. Calculate the mean and standard deviation of the annual emission for the U.S., U.K., Germany, and China. Is this metric useful? In which year was the peak emission from these countries? Are they starting to gain control of their emissions? **30 points**

In order to get the total annual emission of for every country, we must combine the production-based CO2 emission and the consumption-based CO2 emission. To do this, we compute the mean of the "co2" and "consumption_co2" columns after they are added together 

```python
#Create new dataframes to represent the data for each of the mentioned countries.
df_US = df[df.country == 'United States']
df_UK = df[df.country == 'United Kingdom']
df_GER = df[df.country == 'Germany']
df_CHN = df[df.country == 'China']

#Create a new column to the dataframe which adds up the annual production CO2 emission and annual consumption CO2 emission.
df_US['total_co2'] = (df_US.co2+ df_US.consumption_co2.fillna(0))
df_UK['total_co2'] = (df_UK.co2+ df_UK.consumption_co2.fillna(0))
df_GER['total_co2'] = (df_GER.co2+ df_GER.consumption_co2.fillna(0))
df_CHN['total_co2'] = (df_CHN.co2+ df_CHN.consumption_co2.fillna(0))

#Calculate the mean, standard deviation and median while ignoring nan values for USA.
mean_US = np.nanmean(df_US.co2 + df_US.consumption_co2.fillna(0))
std_US = np.nanstd(df_US.co2 + df_US.consumption_co2.fillna(0))
median_US = np.nanmedian(df_US.co2 + df_US.consumption_co2.fillna(0))

#Print the results for United States annual CO2 emission
print('The mean annual CO2 emission for the United States is: ' + str(mean_US) + ' million tonnes.')
print('The standard deviation of annual CO2 emission for the United States is: ' + str(std_US) + ' million tonnes.')
print('The median annual CO2 emission for the United States is: ' + str(median_US) + ' million tonnes. \n')

#Calculate the mean, standard deviation and median while ignoring nan values for UK.
mean_UK = np.nanmean(df_UK.co2 + df_UK.consumption_co2.fillna(0))
std_UK = np.nanstd(df_UK.co2 + df_UK.consumption_co2.fillna(0))
median_UK = np.nanmedian(df_UK.co2 + df_UK.consumption_co2.fillna(0))

#Print the results for United Kingdom annual CO2 emission
print('The mean annual CO2 emission for the United Kingdom is: ' + str(mean_UK) + ' million tonnes.')
print('The standard deviation of annual CO2 emission for the United Kingdom is: ' + str(std_UK) + ' million tonnes.')
print('The median annual CO2 emission for the United Kingdom is: ' + str(median_UK) + ' million tonnes. \n')

#Calculate the mean, standard deviation and median while ignoring nan values for Germany.
mean_GER = np.nanmean(df_GER.co2 + df_GER.consumption_co2.fillna(0))
std_GER = np.nanstd(df_GER.co2 + df_GER.consumption_co2.fillna(0))
median_GER = np.nanmedian(df_GER.co2 + df_GER.consumption_co2.fillna(0))

#Print the results for Germany annual CO2 emission
print('The mean annual CO2 emission for Germany is: ' + str(mean_GER) + ' million tonnes.')
print('The standard deviation of annual CO2 emission for Germany is: ' + str(std_GER) + ' million tonnes.')
print('The median annual CO2 emission for Germany is: ' + str(median_GER) + ' million tonnes. \n')

#Calculate the mean, standard deviation and median while ignoring nan values for China.
mean_CHN = np.nanmean(df_CHN.co2 + df_CHN.consumption_co2.fillna(0))
std_CHN = np.nanstd(df_CHN.co2 + df_CHN.consumption_co2.fillna(0))
median_CHN = np.nanmedian(df_CHN.co2 + df_CHN.consumption_co2.fillna(0))

#Print the results for China annual CO2 emission
print('The mean annual CO2 emission for China is: ' + str(mean_CHN) + ' million tonnes.')
print('The standard deviation of annual CO2 emission for China is: ' + str(std_CHN) + ' million tonnes.')
print('The median annual CO2 emission for China is: ' + str(median_CHN) + ' million tonnes. \n')

#Although it is not required, we can easily visualise this again in abox plot.
boxplot = [df_US['total_co2'],df_UK['total_co2'],df_GER['total_co2'],df_CHN['total_co2']]

plt.boxplot(boxplot)
plt.title('Annual CO2 Emission for the selected Countries')
plt.xlabel('Country')
plt.ylabel('Annual CO2 Emission (million tonnes)')
plt.xticks([1, 2, 3, 4], ['United States','United Kingdom', 'Germany', 'China'])
plt.xticks(rotation=45)
plt.show()
plt.close()
```

This metric is somewhat useful, if one is trying to look at which country has the highest annual emission over the time period. Someone trying to determine the level of annual emission over the entire history could find this useful, but the resulting
value does not allow us to make any predictions as it does not tell us when most of the emissions took place. In our case, we are trying to make a conclusion about if the countries are starting to gain control over their emissions, and this value does not give us much information about that. This is why we find the year with the peak emission as shown below.

```python
#Then, get the year by sorting the column in descending order and taking the top value.
yearUS = df_US.sort_values('total_co2', ascending = False).head(1)['year'].values[0]
print('Year of peak CO2 emission for the United States: ' + str(yearUS))

yearUK = df_UK.sort_values('total_co2', ascending = False).head(1)['year'].values[0]
print('Year of peak CO2 emission for the United Kingdom: ' + str(yearUK))

yearGER = df_GER.sort_values('total_co2', ascending = False).head(1)['year'].values[0]
print('Year of peak CO2 emission for Germany: ' + str(yearGER))

yearCHN = df_CHN.sort_values('total_co2', ascending = False).head(1)['year'].values[0]
print('Year of peak CO2 emission for China: ' + str(yearCHN))
```

Therefore, in order to conclude whether a country is gaining control over their emission, we can make the simplistic assumption that countries with a peak emission further in the past have a better control over their emission, and countries with a more recent peak emission means that they do not have much control over their emission. According to these results, and from the simplistic argument, once could say that Germany has the most control over its emissions, followed by the UK, the US, and finally China, who has the least control over their emission. This argument could be flawed however, as it is possible that one year could have been an outlier and that the general trend is still increasing for a country. Such a result would mean that although the peak emission year was further in the past, the country does not have control over its emission. Such a trend could only be determined if the last ~20-50 years were plotted for every country, but as this exercise does not ask for futher analysis, we will leave it for another time.
